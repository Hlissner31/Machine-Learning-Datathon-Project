```python
import pandas as pd

df = pd.read_csv('aggregation_costreport.csv')
df.describe()
df.info()
```

```python
df.info()
```

```python
# Finding mean revenue
df.groupby('year')['Gross_Revenue'].agg(['mean'])
```

```python
# Finding mean profit
df.groupby('year')['Total_Income'].agg(['mean'])
```

```python
#Finding mean costs (Most important)
df.groupby('year')['Total_Costs'].agg(['mean'])
```

```python
df
```

```python
#Making new dataframe with only texas data. Checking for null values.
tx_data = df[df["State_Code"] == "TX"]
tx_data

tx_county_income = df.groupby('County')['Net_Income'].mean().reset_index()
tx_county_income

tx_county_income.sort_values(by='Net_Income', ascending=False)



tx_data.Net_Income.isna().sum()/len(tx_data)

tx_data.Total_Costs.isna().sum()/len(tx_data)

tx_data.Gross_Revenue.isna().sum()/len(tx_data)


```

```python
#Checking for null values by county
null_counts_by_county = tx_data.groupby('County').apply(lambda x: x.isnull().sum())

print(null_counts_by_county)


```

```python
# Getting a list of all unique counties. Data really needs to be cleaned.
county_count = tx_data['County'].nunique()
county_count

pd.set_option('display.max_rows', None)


unique_counties = tx_data['County'].unique()
unique_counties
unique_counties_sorted = pd.Series(unique_counties).sort_values()
unique_counties_sorted

```
